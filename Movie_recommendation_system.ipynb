# Data cleaning and Preprocessing

```python
# checking for misssing values.

movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
# collecting and joining the genres in a single list

movies['genres'].iloc[0]
```

```python
import ast
```

```python
def convert(text):
    lst =[]
    for i in ast.literal_eval(text):
        lst.append(i['name'])
    return lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.drop('id',axis=1)
```

```python
movies['cast'].iloc[0]
```

```python
#collecting top 3 actors

def convert2(text):
    lst =[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter !=3:
            lst.append(i['name'])
            counter += 1  
            
    return lst
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
# fetching the director from crew
```

```python

def convert3(text):
    l =[]
    
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            l.append(i['name'])
        
    return l
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
movies.head()
```

```python
# removing the spaces in cast,crew ,genres and keyword.

def removingspaces(x):
    l1=[]
    for i in x:
        l1.append(i.replace(" ",""))
        
    return l1
```

```python

movies['cast'] = movies['cast'].apply(removingspaces)
movies['crew'] = movies['crew'].apply(removingspaces)
movies['genres'] = movies['genres'].apply(removingspaces)
movies['keywords'] = movies['keywords'].apply(removingspaces)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['overview'].iloc[0]
```

```python
movies['tags'] = movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['crew']
```

```python
movies_new = movies[['movie_id','title','tags']]
```

```python
movies_new
```

```python
movies_new['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies_new
```

```python
movies_new['tags'].iloc[0]
```

```python
movies_new['tags'].apply(lambda x: x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Converting the tags into vectors

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movies_new['tags']).toarray()
```

```python
vector
```

```python
cv.get_feature_names()
```

# Data Modelling and tesing

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#finding the similarity between movies

similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
movies_new[movies_new['title'] == 'Signed, Sealed, Delivered'].index[0]
```

```python
#finding the similarity of given movie with respect to every other movie and listing it sequentially along with its index

list(enumerate(similarity[4803]))
```

```python
# sorting the list in descending order on the basis of similarity score 

sorted(list(enumerate(similarity[2])) ,reverse = True,key = lambda x:x[1])
```

```python
#function for finding the 5 most similar movies to the given movie

def recommend(movie):
    index = movies_new[movies_new['title'] == movie].index[0]
    movies_list = sorted(list(enumerate(similarity[index])) ,reverse = True,key = lambda x:x[1])
    for i in movies_list[1:6]:
        print(movies_new.iloc[i[0]].title)
  
```

```python
recommend('John Carter')
```
